```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image_origin = mpimg.imread('1.jpg') / 255.0
plt.imshow(image_origin)
```

```python
#cung mot anh nhung khac do sang
a = 0.5
b = 0
image_transform = image_origin * a + b
_, axs = plt.subplots(1,2,figsize=(8,8))
for name, img, ax in zip(["image_origin", "image_transform"], [image_origin, image_transform], axs):
  ax.imshow(img)
  ax.set_title(name)
plt.show()
```

```python
#triet tieu a va b
#triet tieu a -> cosine similarity

def norm2(x, axis = None):
  return np.sqrt(np.sum(x**2, axis= axis, keepdims=True))

image_origin_norm = image_origin / norm2(image_origin, axis = (0,1))
image_transform_norm = image_transform / norm2(image_transform, axis = (0,1))

x = np.abs(image_origin_norm - image_transform_norm).sum()
print(x)
```

```python
#triet tieu b dung mean
a = 1
b = 0.05
image_transform = image_origin * a + b
_, axs = plt.subplots(1,2,figsize=(8,8))
for name, img, ax in zip(["image_origin", "image_transform"], [image_origin, image_transform], axs):
  ax.imshow(img)
  ax.set_title(name)
plt.show()
```

```python
(image_origin - image_transform).sum()
```

```python
image_origin_mean = image_origin - np.mean(image_origin, axis=(0,1))
image_transform_mean = image_transform - np.mean(image_transform, axis=(0,1))

(image_origin_mean - image_transform_mean).sum()
```

```python
#triet tieu ca a va b -> dung correlation = mean + norm
a = 1.5
b = 0.05
image_transform = image_origin * a + b
_, axs = plt.subplots(1,2,figsize=(8,8))
for name, img, ax in zip(["image_origin", "image_transform"], [image_origin, image_transform], axs):
  ax.imshow(img)
  ax.set_title(name)
plt.show()
```

```python
image_origin_mean = image_origin - np.mean(image_origin, axis = (0,1))
image_transform_mean = image_transform - np.mean(image_transform, axis = (0,1))

image_origin_mean_norm = image_origin_mean / norm2(image_origin_mean, axis = (0,1))
image_transform_mean_norm = image_transform_mean / norm2(image_transform_mean, axis = (0,1))

np.sum(image_origin_transform_mean - image_origin_mean_norm)
```
